```python
# calculate the percentage of missing value
null_var = df.isnull().sum()/df.shape[0] *100   #check percentage  shape[0] rows 
null_var
```

```python
# Droping the column/feature whose missing value is more then 17%
drop_columns  = null_var[null_var >17].keys()      #keys column name
drop_columns
```

```python
# Final drop missing value
df2_drop_clm=df.drop(columns=drop_columns)
```

```python
df2_drop_clm.shape
```

```python
# Rechecking the missing value
sns.heatmap(df2_drop_clm.isnull())
```

```python
# droping all the left missing value
df3_drop_rows = df2_drop_clm.dropna()
```

```python
'''
Creating a for loop and then creating plot for each and looking for its distribution
'''


num_var = ['MSSubClass', 'LotArea', 'OverallQual', 'OverallCond',
       'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2',
       'BsmtUnfSF', 'TotalBsmtSF', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'TotRmsAbvGrd', 'Fireplaces',
       'GarageYrBlt', 'GarageCars', 'GarageArea', 'WoodDeckSF', 'OpenPorchSF',
       'EnclosedPorch', '3SsnPorch', 'ScreenPorch', 'PoolArea', 'MiscVal',
       'MoSold', 'YrSold', 'SalePrice']

plt.figure(figsize=(25,25))       
for i,var in enumerate(num_var):
    plt.subplot(9,4,i+1)
    sns.distplot(df[var], bins=20)
    sns.distplot(df3_drop_rows[var], bins=20)

plt.show()
```

From the above we can see that how all the data is distributed and which type of skewness does it followes.
